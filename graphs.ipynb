```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
with open('op.csv', 'r') as f:
    reader = csv.reader(f)
    data = list(reader)

print(data)
```

```python
data_array = np.array(data[1:], dtype=float)
```

```python
print(data_array.shape)
```

```python
t = data_array.T[0]
x = np.abs(data_array.T[1])
y = np.abs(data_array.T[2])
z = np.abs(data_array.T[3])
```

```python
fig, ax = plt.subplots(1, figsize=(12,8))
plt.grid(alpha=0.8)
plt.plot(t, x, label="x")
plt.plot(t, y, label="y")
plt.plot(t, z, label="z")
plt.legend()
ax.set_ylabel(r"Accel (g)", fontsize=15)
ax.set_xlabel(r"Time (ms)", fontsize=15)
```

```python
fig, ax = plt.subplots(1, figsize=(12,8))
plt.grid(alpha=0.8)
plt.plot(t[10:120], x[10:120], label="x")
plt.plot(t[10:120], y[10:120], label="y")
plt.plot(t[10:120], z[10:120], label="z")
plt.legend()
ax.set_ylabel(r"Accel (g)", fontsize=15)
ax.set_xlabel(r"Time (ms)", fontsize=15)
```

# SENSOR WITH KNOCKS

```python
with open('op_knocks.csv', 'r') as f:
    reader = csv.reader(f)
    data = list(reader)

print(data)
```

```python
data_array = np.array(data[1:], dtype=float)
```

```python
print(data_array.shape)
```

```python
t = data_array.T[0] - data_array.T[0][0]
x = data_array.T[1] 
x2= x* 1.0/4096
```

```python
fig, ax1 = plt.subplots(1, figsize=(6,4))
plt.grid(True, alpha=0.4)
ax1.set_xlabel(r"Time (ms)", fontsize=15)
ax1.set_ylabel(r"Accel (g)", fontsize=15)
ax1.plot(t, x2)
ax1.tick_params(axis='y')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.set_ylabel(r"Accel (raw)", fontsize=15)  # we already handled the x-label with ax1
ax2.yaxis.set_ticks(np.arange(0, 10000, 2048))
ax2.plot(t, x)
ax2.tick_params(axis='y')

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title("Absolute Acceleration w/ Knocks")
```

```python
with open('op_knocks_even.csv', 'r') as f:
    reader = csv.reader(f)
    data = list(reader)

print(data)
```

```python
data_array = np.array(data[1:], dtype=float)
```

```python
print(data_array.shape)
```

```python
ts = data_array.T[0] - data_array.T[0][0]
xs = data_array.T[1] 
```

```python
last_knock = 0
knk_db = 80
knock_times = []
for x,t in zip(xs,ts):
    if x > 1400 and t - last_knock > knk_db:
        last_knock = t
        knock_times.append(last_knock)
knock_times = np.array(knock_times)
knock_times
```

```python
fig, ax1 = plt.subplots(1, figsize=(6,4))
plt.grid(True, alpha=0.4)
ax1.set_xlabel(r"Time (ms)", fontsize=15)
ax1.set_ylabel(r"Accel (4096 * g)", fontsize=15)
ax1.plot(ts, xs)
ax1.scatter(knock_times, 1500*np.ones(len(knock_times)), color ="r")
ax1.tick_params(axis='y')

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title("Thresholding Test")

```

```python
time_diffs = np.diff(knock_times)
print(np.std(time_diffs))
```
